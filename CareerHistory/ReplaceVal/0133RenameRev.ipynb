```python
def dict_update_educ(x, y, dict_original):
    # Automate renaming for q26 and q28
    
    dict_new = {
        x + '_1': y + '_deg',
        x + '_1sq1': y + '_hs_maj',
        x + '_1sq2': y + '_clg_maj',
        x + '_2': y + '_pref',
        x + '_3': y + '_est',
        x + '_4': y + '_crs',
        x + '_5': y + '_grad',
        x + '_6_1': y + '_s_age',
        x + '_6_2': y + '_e_age',
        }
    dict_original.update(dict_new)
    
    return dict_original    
    
```

```python
def dict_update_chng(x, y, dict_original):
    # Automate renaming for q16
    
    dict_new = {
        x + '_kigyou': y + '_frm_id',
        x + '_jigyousho': y + '_est_id',
        x + '_1': y + '_emp_pos',
        x + '_3': y + '_pref',
        x + '_3_toshiruikei': y + '_ctype',
        x + '_4': y + '_ind',
        x + '_5_dai': y + '_ind_maj',
        x + '_5_shou': y + '_ind_min',
        x + '_6': y + '_pos',
        x + '_7': y + '_f_sz',
        x + '_8': y + '_chnl',
        x + '_9_1': y + '_s_age',
        x + '_9_2': y + '_e_age',
        x + '_10' : y + '_chng_rsn',
        }
    dict_original.update(dict_new)
    
    return dict_original
```

```python
def dict_update_jobs(x, y, dict_original):
    # Automate renaming for q6
    
    dict_new = {
        x + '_kigyou': y + '_frm_id',
        x + '_jigyousho': y + '_est_id',
        x + '_1': y + '_emp_pos',
        x + '_3': y + '_pref',
        x + '_3_toshiruikei': y + '_ctype',
        x + '_4': y + '_ind',
        x + '_5_dai': y + '_ind_maj',
        x + '_5_shou': y + '_ind_min',
        x + '_6': y + '_pos',
        x + '_7': y + '_f_sz',
        x + 'a': y + '_chnl',
        x + 'b': y + '_s_age',
        x + 'c': y + '_e_age'
        }
    dict_original.update(dict_new)
    
    return dict_original
```
