```python
def slice_df(condition):
    _df = df[condition]
    _df = _df[['init_ind_cl','curr_ind_cl','chrt']]
    _df['same'] = np.where(_df.curr_ind_cl == _df.init_ind_cl, True, False).astype(int)
    _df['total'] = 1 # use ones to create total
    _df_dmmy = pd.get_dummies(_df, columns = ['curr_ind_cl', 'init_ind_cl', 'same'])
    _df_chrt = _df_dmmy.groupby('chrt').sum().reset_index()
    _df_chrt['class'] = _df_chrt['chrt'] + 15
    return _df_chrt
```

```python
#_df_chrt.to_csv('male_ind.csv')
```

```python
_df_chrt = slice_df(df['init_j_edu_deg']==7)
_df = _df_chrt.set_index('class')
fig, ax = plt.subplots()
#ax = _df.total.plot(kind = 'bar', color = 'b', label = 'total')
ax = _df.same_0.plot(kind = 'bar', color = 'r', label = 'No')
ax = _df.same_1.plot(kind = 'bar', bottom = _df.same_0, 
                                                   color = 'g', label ='Yes')

plt.title('Is curr ind the same as init ind?')
ax.set_xlabel('Class')
ax.set_ylabel('Counts')
plt.legend()
plt.show()
```

```python
_df_chrt = slice_df(df['init_j_edu_deg']==7)
_df = _df_chrt.set_index('class')

_df['same_1_shr'] = _df.same_1/_df.total
_df['same_0_shr'] = _df.same_0/_df.total

_df_1 = _df.same_1_shr
_df_2 = _df.same_0_shr
fig, ax = plt.subplots()
#ax = _df.total.plot(kind = 'bar', color = 'b', label = 'total')
ax = _df_1.plot(kind = 'bar', color = 'r', label = 'Yes')
ax = _df_2.plot(kind = 'bar', bottom = _df_1, color = 'g', label ='No')

plt.title('Is curr ind the same as the init ind?')
ax.set_xlabel('Class')
ax.set_ylabel('Share')
plt.legend()
plt.show()
```
