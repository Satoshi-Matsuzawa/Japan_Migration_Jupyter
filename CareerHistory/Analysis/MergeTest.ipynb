```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
df_male = pd.read_csv('male.csv', index_col = 0)
df_female = pd.read_csv('female.csv', index_col = 0)
```

```python
df_male
```

```python
df = df_male.merge(df_female, left_on = 'class', right_on = 'class' , how = 'inner')
```

```python
keywords = ['AD', 'SCH','ORS', 'SE','CONN', 'UNK', 'total']
```

```python
for keyword in keywords:
    column = df.filter(regex = (keyword)).columns
    df[keyword] = df[column[0]] + df[column[1]]
```

```python
df.filter(regex=('AD')).columns[0]
```

```python
df
```

```python
_df = df.set_index('class')
fig, ax = plt.subplots()
ax = _df.total.plot(kind = 'bar', color = 'b', label = 'Total')
ax = _df.SCH.plot(kind = 'bar', color = 'r', label = 'School')
ax = _df.CONN.plot(kind = 'bar', bottom = _df.SCH, color = 'g', label ='Connection')
ax = _df.SE.plot(kind = 'bar', bottom = _df.SCH + _df.CONN, color = 'y', label ='Self Employed')



plt.title('Employment channel for the first job')
ax.set_xlabel('Class')
ax.set_ylabel('Counts')
plt.legend()
plt.show()
```
