```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file = '../../../Data/Workers/1207/1207.csv'
df = pd.read_csv(file, low_memory = False)
```

```python
df['counts'] = 1
```

```python
area_str = df.area12y.unique().astype(str).tolist()
```

```python
tokyo_str = [item for item in area_str if item.startswith('13')]
```

```python
tokyo_int = [int(item) for item in tokyo_str]
```

```python
df[df.area12y.isin(tokyo_int)]
```

```python
kanagawa_str = [item for item in area_str if item.startswith('14')]
```

```python
kanagawa_int = [int(item) for item in kanagawa_str]
```

```python
df[df.area12y.isin(kanagawa_int)]
```

```python
df_area = df[['counts','area12y']].groupby('area12y').sum().reset_index()
```

```python
df_area
```

```python
df_area.counts
```

```python
fig, ax = plt.subplots()
ax.plot(df_area.area12y, df_area.counts)
```
