```python
def dict_update_educ(x, y, dict_original):
    # Automate renaming for q26 and q28
    
    dict_new = {
        x + '_1': y + '_bkg',
        x + '_1sq1': y + '_hs_maj',
        x + '_1sq2': y + '_clg_maj',
        x + '_2': y + '_pref',
        x + '_3': y + '_est',
        x + '_4': y + '_course',
        x + '_5': y + '_grad',
        x + '_6_1': y + '_sage',
        x + '_6_2': y + '_eage',
        }
    dict_original.update(dict_new)
    
    return dict_original    
    
```

```python
def dict_update_trans(x, y, dict_original):
    # Automate renaming for q16
    
    dict_new = {
        x + '_kigyou': y + '_job_emp_pos',
        x + '_jigyousho': y + '_job_pref',
        x + '_1': y + '_job_emp_pos',
        x + '_3': y + '_job_pref',
        x + '_3_toshiruikei': y + '_job_ctype',
        x + '_4': y + '_job_ind',
        x + '_5_dai': y + '_job_ind_maj',
        x + '_5_shou': y + '_job_ind_min',
        x + '_6': y + '_job_pos',
        x + '_7': y + '_job_fsize',
        x + '_8': y + '_job_chnl',
        x + '_9_1': y + '_job_sage',
        x + '_9_2': y + '_job_eage',
        x + '_10' : y + '_job_chng_rsn'
        }
    dict_original.update(dict_new)
    
    return dict_original
```

```python
def dict_update_jobs(x, y, dict_original):
    # Automate renaming for q6
    
    dict_new = {
        x + '_1': y + '_job_emp_pos',
        x + '_3': y + '_job_pref',
        x + '_3_toshiruikei': y + '_job_ctype',
        x + '_4': y + '_job_ind',
        x + '_5_dai': y + '_job_ind_maj',
        x + '_5_shou': y + '_job_ind_min',
        x + '_6': y + '_job_pos',
        x + '_7': y + '_job_fsize',
        x + 'a': y + '_job_chnl',
        x + 'b': y + '_job_sage',
        x + 'c': y + '_job_curr_age'
        }
    dict_original.update(dict_new)
    
    return dict_original
```

```python
list_educ = [('q26','init_job_educ'),
             ('q28', 'final_educ')]
```

```python
for item in list_educ:
    dict_update_educ(item[0],item[1],rename_col)
    
```

```python
list_jobs = [('q6','curr')]
```

```python
for item in list_jobs:
    dict_update_jobs(item[0],item[1],rename_col)
```
