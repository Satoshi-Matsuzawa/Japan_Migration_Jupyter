```python
def plot_histogram(df, column_name, bins=20):
    sns.set()
    _ = plt.hist(df[column_name], bins= bins)
    _ = plt.xlabel(column_name)
    _ = plt.ylabel('counts')
    _ = plt.title('Histgram of ' + column_name)
    plt.show()
    
plot_histogram(df, 'male_graduates')
```
