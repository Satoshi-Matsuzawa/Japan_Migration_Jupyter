```python
df2=pd.get_dummies(df, columns=['employment_channel']).groupby('school_id').sum()
```

```python
df_channels['school_and_shokuan']=df_channels.iloc[:,0]+df_channels.iloc[:,1]+df_channels.iloc[:,2]
```

```python
df_channels['family_and_connection']=df_channels.iloc[:,3]+df_channels.iloc[:,4]+df_channels.iloc[:,5]
```

```python
df_channels = df_channels.iloc[:,-2:].reset_index()
```

```python
df_school = pd.read_csv('../../../../Data/School/0001/school_cleaned.csv')
```

```python
df_school
```

```python
df_merged = pd.merge(df_school, df_channels, on='school_id')
```

```python
df_merged
```

```python
df_merged['total_employed']=df_merged['male_employed']+df_merged['female_employed']
```

```python
df_merged['shokuan_share']=df_merged['school_and_shokuan']/df_merged['total_employed']
```

```python
df_merged['connection_share']=df_merged['family_and_connection']/df_merged['total_employed']
```

```python
df_merged['total_graduates']=df_merged['male_graduates']+df_merged['female_graduates']
```

```python
df_merged.describe()
```

```python
def create_share(df, gender, category):
    df[gender+ category+'_share']=df[gender+ category]/df[gender+'_graduates']
```

```python
create_share(df_merged, 'male','_high_school')
```

```python
import numpy as np
```

```python
# Hypothesis: the first two digits in the school_id represents the city
school_group = [item[:-1] for item in df.school_id.unique().astype(str)]
school_group = np.unique(np.array(school_group).astype(int))

# Add school groups to the dataframe
for item in school_group:
    search = item.astype(str) + '[0-9]$'
    bool_match = df.school_id.astype(str).str.match(search)
    df.loc[bool_match, 'school_group'] = item
```
