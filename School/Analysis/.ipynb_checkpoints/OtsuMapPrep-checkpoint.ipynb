```python
def apply_dict(col, dictionary, df = df):
    df[col] = df[col].map(dictionary)
```

```python
col_dict = [['seibetu', gndr],
            ['q3.1', heir],
            ['q3.2', gdn_rel],
            ['q4.1', gdn_occ],
            ['q6', ind],
            ['q7', lctn],
            ['q12', ch],
            ['q11', comm],
           ]
```

```python
for item in col_dict:
    col = item[0]
    dictionary = item[1]
    apply_dict(col, dictionary)
```

```python
col_names = {
    'seibetu' : 'gndr',
    'q3.1' : 'heir',
    'q3.2' : 'gdn_rel',
    'q4.1' : 'gdn_occ',
    'q4.3.2': 'fam_sz',
    'q6': 'ind',
    'q7': 'lctn',
    'q8': 'frm_sz',
    'q10': 'sal',
    'q11': 'comm',
    'q12': 'ch',
}
```
